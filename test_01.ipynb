```python
#for문 : for 변수명 in range(시작값,종료값,증가값)
#range() -> 시작값부터 시작,증가값만큼 증가, 종료값에 종료하나 종료값 포함x
#           반복 횟수/숫자 인덱스용 시퀀스 생성기
#           range 없이 쓰면 컬렉션의 원소 자체를 하나씩 꺼낸다

print(
    list(range(1,10,1))
)
```

```python
#특정공간(변수를 생성)에 초기값(0)을 지정하고 
result = 0
# 1부터 10까지의 데이터를 반복하여 초기값에 누적합
#1~10이면 종료값은 11을 써야
for i in range(1,11,1):
    
    #i에는 1부터 10까지의 데이터가 1씩 증가하면서 대입
    #i를 result와 더하고 결과를 result에 대입
    result += i

print(result)
```

```python
#while문
#초기값 지정 (시작값 지정)

i=0
while (i <= 10):
#i가 10보다 작거나 같은 경우에 반복 실행
    print(i)
    
    i = i+1
```

```python
#while문을 이용하여 1부터 10까지의 합계 구하기

#초기값 지정
i = 0
result = 0

while (i <= 10):
    result = result + i
    #처음에 i=0이기때문에 영원히 while조건식이 참이다
    # -> i= i+1을 해줌으로써 i가 10이하까지 반복하게 만들고 
    # 반복문 종료하게만듬


    i = i+1
print(result)
```

```python
#반복문 안에서 조건식을 이용하여 break를 이용
# 1부터 시작해서 1씩 증가시킨 데이터를 누적합하였을때
# 누적합의 값이 7000이 넘어가는 시점은 언제인가? 이 결과를 확인하기위해서 가장 간단한 것 부터 하나씩 확인하면서 결과 도출하는 연습 필요
# -> result가 7000이상이되는 조건식 생성
# case 1 (for문)
# 누적합을 해줄 공간을 생성

result = 0

for i in range(1,1000,1):
    result += i

    if result > 7000:
        print(i)
        break
    
print(result)
```

```python
#반복문 안에서 조건식을 이용하여 break를 이용
# 1부터 시작해서 1씩 증가시킨 데이터를 누적합하였을때
# 누적합의 값이 7000이 넘어가는 시점은 언제인가? 이 결과를 확인하기위해서 가장 간단한 것 부터 하나씩 확인하면서 결과 도출하는 연습 필요
# -> result가 7000이상이되는 조건식 생성
# case 2 (while문)
# 누적합을 해줄 공간을 생성

i=0
result=0

while (result < 7000):
    result += i
    i += 1

print(result)
print(i)


```
